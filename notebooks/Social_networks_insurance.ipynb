이중 차이 추정, 제한화 추정, 역시 추정은 "causalinference" 라이브러리에서 제공하는 인과 추론 방법 중 일부입니다. 각각의 추정 방법에 대해 자세히 설명해드리겠습니다:

1. 이중 차이 추정 (Double Difference Estimation):
   이중 차이 추정은 개입 그룹과 비개입 그룹 간의 차이를 시간의 흐름에 따라 비교하는 방법입니다. 주로 처리 그룹과 비처리 그룹의 전후 비교를 통해 개입의 효과를 추정하는 데 사용됩니다. 이 방법은 개입 전과 개입 후, 그리고 처리 그룹과 비처리 그룹 사이의 차이를 비교함으로써 개입의 인과적 효과를 추론합니다.

2. 제한화 추정 (Propensity Score Estimation):
   제한화 추정은 처리 그룹과 비처리 그룹 간의 편향을 줄이기 위해 개입 확률을 고려하는 방법입니다. 개입 확률은 개별 개체가 처리 그룹에 속할 확률을 나타내며, 이를 이용하여 개입 그룹과 비개입 그룹 간의 차이를 보정합니다. 제한화 추정은 개입이 발생할 확률에 따라 그룹을 매칭하거나 가중치를 적용하여 인과 추론을 수행합니다.

3. 역시 추정 (Instrumental Variable Estimation):
   역시 추정은 처리 변수와 결과 변수 사이에 독립적인 도구 변수를 사용하여 인과적 효과를 추정하는 방법입니다. 도구 변수는 처리 변수와 결과 변수 간의 연관성을 갖지만 결과 변수에 직접적으로 영향을 미치지 않는 변수입니다. 역시 추정은 처리 변수의 엔도겐성(endogeneity) 문제를 해결하기 위해 사용됩니다. 이 방법은 도구 변수의 독립성을 활용하여 처리 변수의 영향을 추정하고 결과 변수에 대한 인과적 효과를 분리합니다.

위의 추정 방법들은 "causalinference" 라이브러리에서 제공되는 일부 인과 추론 기법입니다. 각각의 방법은 다른 상황에서 사용되며, 연구의 목적과 데이터의 특성에 따라 선택됩니다. 이러한 방법들은 개입의 인과적 효과를 추정하고 편향을 보정하여 더 신뢰할 수 있는 결과를 도출하는 데 도움을 줍니다.

---

처리 효과 추정과 기계학습 기반 인과 추론은 "causalinference" 라이브러리에서 제공하는 두 가지 다른 방법론입니다. 이들의 주요 차이점을 설명해드리겠습니다:

1. 처리 효과 추정 (Treatment Effect Estimation):
   처리 효과 추정은 개입(처리)의 인과적 효과를 추정하는 데 중점을 둡니다. 이 방법은 개입을 받은 그룹과 비개입 그룹 간의 차이를 분석하여 개입의 효과를 측정합니다. 주로 이중 차이 추정 및 제한화 추정과 관련된 방법을 사용합니다. 처리 효과 추정은 개입 그룹과 비개입 그룹 사이의 차이를 인과적으로 분석하고, 개입의 효과 크기와 통계적 유의성을 추정합니다.

2. 기계학습 기반 인과 추론 (Machine Learning-Based Causal Inference):
   기계학습 기반 인과 추론은 인과적 효과를 추론하기 위해 기계학습 모델을 활용합니다. 이 방법은 개입 변수와 결과 변수 사이의 복잡한 관계를 모델링하고, 개입의 영향을 예측하는 데 기계학습 알고리즘을 사용합니다. 예를 들어, 회귀 분석, 의사결정 트리, 랜덤 포레스트, 신경망 등 다양한 기계학습 모델을 활용하여 인과 관계를 추론할 수 있습니다.

처리 효과 추정은 처리 그룹과 비처리 그룹의 차이를 통해 개입의 효과를 분석하는 반면, 기계학습 기반 인과 추론은 모델을 사용하여 개입의 효과를 예측하고 추론합니다. 처리 효과 추정은 일반적으로 명시적인 모델 가정을 기반으로 하며, 개입 변수와 결과 변수 간의 관계를 통계적으로 분석합니다. 반면에 기계학습 기반 인과 추론은 모델을 사용하여 더 유연하고 복잡한 관계를 모델링하고 예측합니다. 이를 통해 개입의 효과를 예측하고 변수 간의 복잡한 상호작용을 고려할 수 있습니다.

두 방법은 연구의 목적과 데이터의 특성에 따라 선택되며, 개입의 효과를 추정하고 인과적 효과를 이해하는 데 다

---

부분적 자료 처리(partial data treatment)는 인과 추론에서 사용되는 방법 중 하나입니다. 이 방법은 일부 개체 또는 그룹만이 처리(개입)를 받은 경우에 대한 인과 효과를 추정하는 데 사용됩니다.

보통 인과 추론에서는 처리 그룹과 비처리 그룹 사이의 차이를 분석하여 개입의 효과를 추정합니다. 그러나 모든 개체나 그룹이 처리를 받지 않는 경우 또는 처리를 받은 개체 또는 그룹의 수가 제한적인 경우에는 인과 추론이 어려울 수 있습니다. 이때 부분적 자료 처리는 처리를 받은 개체 또는 그룹 중에서 일부만을 사용하여 인과 효과를 추정하는 방법입니다.

부분적 자료 처리는 제한화 추정, 역시 추정 등과 연계하여 사용될 수 있습니다. 예를 들어, 제한화 추정에서는 개입 확률을 고려하여 처리 그룹과 비처리 그룹을 매칭하거나 가중치를 적용합니다. 이때 처리 그룹 중에서 일부만이 개입을 받은 경우에는 이러한 매칭 또는 가중치 적용에서 해당 개체 또는 그룹만을 고려하여 부분적 자료 처리를 수행할 수 있습니다.

부분적 자료 처리는 실제 상황에서 모든 개체 또는 그룹에 개입을 적용하기 어려운 경우에 유용하게 사용될 수 있습니다. 처리를 받은 개체 또는 그룹의 수가 제한적인 경우에도 인과 효과를 추정할 수 있으며, 연구의 효율성과 실용성을 높일 수 있습니다. 그러나 부분적 자료 처리는 선택 편향이나 외생성 가정 등에 영향을 받을 수 있으므로 신중하게 적용해야 합니다.

개체 또는 그룹은 인과 추론에서 사용되는 단위입니다. 이해하기 쉽게 예를 들어 설명하겠습니다.

개체 (Unit):
개체는 인과 추론에서 분석의 대상이 되는 개별적인 단위입니다. 예를 들어, 교육 정책의 효과를 연구한다고 가정해봅시다. 이 경우, 개체는 학생으로 정의될 수 있습니다. 각 학생은 개별적인 개체이며, 개별적인 특성과 결과를 가지고 있습니다. 개체는 보통 독립적으로 측정되고 분석되는 단위입니다.

그룹 (Group):
그룹은 개체들의 집합으로 구성됩니다. 그룹은 개체들을 특정한 특성이나 조건에 따라 구분하여 분석하는데 사용됩니다. 예를 들어, 교육 정책의 효과를 연구한다면, 그룹은 처리 그룹과 비처리 그룹으로 나뉘어질 수 있습니다. 처리 그룹은 특정 교육 개입을 받은 학생들의 집합이고, 비처리 그룹은 해당 개입을 받지 않은 학생들의 집합입니다. 그룹은 일반적으로 개체들을 묶어서 비교하고 인과 효과를 추론하는 데 사용됩니다.

개체와 그룹은 인과 추론에서 중요한 개념으로, 개별적인 개체의 특성과 처리 여부, 그리고 개체를 그룹화하여 처리 그룹과 비처리 그룹을 구분하여 분석합니다. 이를 통해 개체 수준의 효과와 그룹 수준의 효과를 추론할 수 있습니다. 개체와 그룹은 연구의 대상에 따라 다르게 정의될 수 있으며, 연구 질문과 목적에 따라 적절하게 선택되고 사용됩니다.

```python
## 지울 예정...

# install.packages('tidyverse')
# install.packages('modelsummary')
# install.packages('fixest')
# install.packages('causaldata')
# install.packages('gmm')
# install.packages('GGally')
# install.packages('CBPS')
# install.packages('Rtools')
# install.packages('ggpubr')

# Import library
library(tidyverse); library(modelsummary); library(fixest); library(gmm); library(ggplot2); library(GGally)
library(CBPS); 

# import social insure dataset from causaldata package
d <- causaldata::social_insure

# data set describe
head(d)
summary(d) # 데이터 중 age, agpop 등 5개 컬럼에서 NA 값들이 발견
boxplot(d)

# remove NA cell
d_prep <- d %>%
  na.omit()

## data description (KOR)
# - **address** Natural village: 주소
# - **village** Administrative village: 행정 주소 (전처리)
# - **takeup_survey** Whether farmer ended up purchasing insurance. (1 = yes)
# : 최종적인 농부의 보험 가입 여부
# - **age** Household Characteristics - Age: 가구특성-연령
# - **agpop** Household Characteristics - Household Size: 가구특성-가구 규모
# - **ricearea_2010** Area of Rice Production: 쌀 생산 면적 (농부의 자산, 생산 규모)
# - **disaster_prob** Perceived Probability of Disasters Next Year: 차년도 재해 발생 확률
# - **male** Household Caracteristics: Gender of Household Head (1 = male): 가구주 성별 (1 남성)
# - **default** "Default option" in experimental format assigned to. (1 = default is to buy, 0 = default is to not buy): 실험 양식 할당 번호: 기본값=1은 구매(보험 가입), 0은 미구매
# - **intensive** Whether or not was assigned to "intensive" experimental session (1 = yes)
# : 실험 세션에 할당되었는지 여부
# - **risk_averse** Risk aversion measurement: 위험 회피 성향 점수
# - **literacy** 1 = literate, 0 = illiterate: 문해력, 1=읽고쓸줄 앎, 0=문맹
# - **pre_takeup_rate** Takeup rate prior to experiment: 실험 전 보험 수급율

# 산포도 및 변수간 상관성 시각화
# ggpairs( d %>%
#            select(takeup_survey, male, age, agpop, ricearea_2010,
#                   literacy, intensive, risk_averse, disaster_prob,
#                   pre_takeup_rate, default))

###############################################

# 실험 참여 여부를 -> treatment로 지정
d_prep$treatment<-d_prep$intensive

# treat 0,1 숫자 확인
table(d_prep$treatment)

# 0   1 
# 706 672 

# d_prep$treatment <- as.factor(d_prep$treatment) # 0,1 숫자값을 factor로 변경

#calculate propensity score using logistic regression model
prop_model <-glm(treatment~takeup_survey+male+age+agpop+ricearea_2010+
                 literacy+risk_averse+disaster_prob+
                 village+address+pre_takeup_rate,data=d_prep,family=binomial(link="logit"))

#save predicted scores to dataset
d_prep$pr_score <-predict(prop_model,type="response")

library(ggplot2)
library(ggpubr)
#this plots the distribution of our estimated propensity scores
g<- ggplot(d_prep,aes(x = pr_score, color=as.factor(d_prep$treatment),fill=as.factor(d_prep$treatment))) +
  geom_density(alpha=.47) + 
  xlab("Estimated Probability of Recieving Oral Therapy") +
  ylab ("Density") +theme_minimal()+ theme(
    axis.ticks.y=element_blank(),
    panel.grid.minor=element_blank(),
    legend.title=element_blank(),
    text = element_text(size=16),
    axis.title.x =element_text(hjust = 0.2,size=16))
ggpar(g,palette="nejm")


library(CBPS)
#calculate weights using CBPS package and function
cbpsoral <-CBPS(treatment~takeup_survey+male+age+agpop+ricearea_2010+
                  literacy+risk_averse+disaster_prob+
                  village+address+pre_takeup_rate,data=d_prep,standardize=FALSE,method="exact",ATT=1)

d_prep$cbps <-cbpsoral$weights

#create matched dataset based on same propensity model
#here we switched the outcome, as matchit requires the larger group to be the reference if variable ratio is being used
#We are capturing up to 4 oral therapy matches for every sipuleucel-T subject
matched <- matchit((1-treatment)~takeup_survey+male+age+agpop+ricearea_2010+
                     literacy+risk_averse+disaster_prob+
                     village+address+pre_takeup_rate,data =d_prep, method = "nearest",caliper=.2,ratio=4)

#looked at characteristics of matched object
matched_sum<-summary(matched)
# treat = control 그룹 수가 일치하지 않기에 일부 treat 그룹은 매칭을 받지 못하였다.
# Warning message: Fewer control units than treated units; not all treated units will get a match. 
```
